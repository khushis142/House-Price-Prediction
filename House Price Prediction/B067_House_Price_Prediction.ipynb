# Cleaning the data

Columns to drop:

*   We drop country because it has only 1 unique value
*   We drop date because all were built in the same year and date does not show much correlation with price
*   We drop view, waterfront because most values in it are 0
*   We drop street and because it has too many unique values and does not show a trend

```python
df.drop(['country', 'date', 'view', 'waterfront', 'street','statezip'], axis=1, inplace=True)
```

```python
#df.drop(['statezip'], axis=1, inplace=True)
```

```python
for i in ['country', 'date', 'view', 'waterfront', 'street']:
  if i in numerical:
    numerical.remove(i)
  elif i in categorical:
    categorical.remove(i)
```

```python
df.head()
```

Checking null values:
* In sqft_basement 0 indicates that there is no basement present.
* In yr_renovated 0 indicates that house has not been renovated we will replace this with yr_built value
* In all other columns 0 indicates null values

```python
for i in df.columns:
  if i not in ['sqft_basement','yr_renovated']:
    df[i].replace(0,np.nan, inplace=True)
```

```python
df.isnull().sum()
```

Not too many columns have null values so we will drop the null values

```python
df=df.dropna()
```

```python
df.reset_index(inplace=True)
```

```python
l=[]
for i in range(len(df['yr_renovated'])):
  if df['yr_renovated'][i]== 0:
    l.append(df['yr_built'][i])
  else:
    l.append(df['yr_renovated'][i])
df['yr_renovated']=l
df.head()
```

```python
df.drop(['index'], axis=1, inplace=True)
```

Checking for duplicates:

```python
duplicate = df[df.duplicated(keep = 'last')]
 
print("Duplicate Rows :")
 
# Print the resultant Dataframe
duplicate
```

```python
df.info()
```

There are no duplicates

Removing outliers:


```python
for i in numerical:
  for x in [i]:
    q75,q25 = np.percentile(df.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    df.loc[df[x] < min,x] = np.nan
    df.loc[df[x] > max,x] = np.nan
```

# Comparing Models

We will compare:

* Linear regression
* Lasso regression
* Ridge regression
* Decision tree regressor
* Random Forest regressor
* Extra Tree regressor
* Extreme Gradient Boosting

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor 
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import ExtraTreesRegressor
import xgboost as xg
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
```

```python
df2=pd.get_dummies(df)
```

```python
df2.shape
```

```python
X_head = df2.iloc[:, df2.columns != 'price']
X = df2.loc[:, df2.columns != 'price']
y = df2['price']
X_head.head()
#X_head.drop('index', axis=1, inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.20, random_state=0)
```

```python
MLA_test = []
MLA_train=[]
Z = [LinearRegression(), Lasso(alpha=1), Ridge(alpha=1), DecisionTreeRegressor(max_depth=10, min_samples_leaf=3, min_samples_split=3), RandomForestRegressor(max_depth=10, min_samples_split=4, min_samples_leaf=4), xg.XGBRegressor(max_depth=4,min_child_weight=1.5)]
X = ["Linear Regression", "Ridge Regression", "Lasso Regression", "Decision Tree Regressor", "Random Forest Regressor", "Extreme Gradient Boosting"]

for i in range(0,len(Z)):
    model = Z[i]
    model.fit( X_train , y_train )
    #pred = model.predict(X_test)
    MLA_test.append(model.score(X_test , y_test))
    MLA_train.append(model.score(X_train , y_train))
```

```python
d = { "Algorithm" : X, "R2 score train" : MLA_train, "R2 score test": MLA_test }

dfm = pd.DataFrame(d)
dfm
```

```python
model = xg.XGBRegressor()
model.fit( X_train , y_train )
feat_importances = pd.Series(model.feature_importances_, index=X_head.columns)
feat_importances.nlargest(25).plot(kind='barh',figsize=(10,10))
```
